```python
def retrieveNearByRest(lat,lng,radius,idx):
    
    try:
        places_result = gmaps.places_nearby(location= f'{lat},{lng}',radius = radius,
                                open_now = False, type = 'restaurant')
        first_20 = places_result['results']
    except:
        print("first twenty place gone south")
        return False,[]
    time.sleep(3)
    try:
        next_page = gmaps.places_nearby(page_token = places_result['next_page_token'])
        latter_20 = next_page['results']
    except:
        print("LATTER twenty place gone south")
        return False,[]
    restDict = {}
    for restaurant in first_20+latter_20:
        try:
            if restaurant['user_ratings_total']<=20:
                continue
        except:
            print(f"skipping {restaurant['name']} due to lack of ratings")
            continue
        if 'price_level'  not in restaurant:
            price_level = 'N/A'
        else:
            price_level = restaurant['price_level']
        restDict[restaurant["place_id"]] = {'rating':restaurant['rating'],
                                            'price_level':price_level,
                                            'lat':restaurant['geometry']['location']['lat'],
                                            'lng':restaurant['geometry']['location']['lng'],
                                            'type':'restaurant',
                                            'name':restaurant['name'],
                                            'vicinity':restaurant['vicinity'],
                                            'affili':idx
                                           }

    return True,restDict
        
        
```

```python
# retrieving the top 40 most prominent restaurant around each location
nearBy_restaurant = {}
potential_work_latlng = []
headers = ['name','address','type','rating','lat','lng','price_level','place_id','affili']
# near by places within 10 miles radius
radius = 16093
for idx in range(1,len(loc_data.dropna())):
    try:
        tmp_geo = gmaps.geocode(loc_data.loc[idx]['Address'])
    except:
        print("tmp geo code wrong")
    lat = tmp_geo[0]['geometry']['location']['lat']
    lng = tmp_geo[0]['geometry']['location']['lng']
    potential_work_latlng.append({'name':loc_data.loc[idx]['Name'],
                                  'address':loc_data.loc[idx]['Address'],
                                  'lat':lat,
                                  'lng':lng
                                 })
    status, struc = retrieveNearByRest(lat,lng,radius,idx)
    if status:
        nearBy_restaurant.update(struc)
    else:
        print(f"this data has gone wrong: {loc_data.loc[idx]}")
    
    
    
    
    
```

```python
# due to the limited nature of the events that came up between
# May 16 2022 and July 22 2022
# I manually import some of the event data from the referenced website to a excel sheet

# combine with event that is not restricted by date: i.e. location-based events
# amusment park, bowling alley can serve for gaming

# gym can serve for sport event

# art gallery and museum can serve for art event

def biweeklyEvent(lat,lng,radius,eventType,idx):
    
    try:
        places_result = gmaps.places_nearby(location= f'{lat},{lng}',radius = radius,
                                open_now = False, type = eventType)
        first_20 = places_result['results']
    except:
        print("first twenty place gone south")
        return False,[]
    
    
    restDict = {}
    for restaurant in first_20:
        try:
            if restaurant['user_ratings_total']<=20:
                continue
        except:
            print(f"skipping {restaurant['name']} due to lack of ratings")
            continue
        if 'price_level'  not in restaurant:
            price_level = 'N/A'
        else:
            price_level = restaurant['price_level']
        if eventType == 'gym':
            outputType = 'sport'
        elif eventType == 'art_gallery' or eventType == 'museum':
            outputType = 'art'
        else:
            outputType = 'gaming'
        restDict[restaurant["place_id"]] = {'rating':restaurant['rating'],
                                            'price_level':price_level,
                                            'lat':restaurant['geometry']['location']['lat'],
                                            'lng':restaurant['geometry']['location']['lng'],
                                            'type':outputType,
                                            'name':restaurant['name'],
                                            'vicinity':restaurant['vicinity'],
                                            'affili':idx
                                           }

    return True,restDict
```

```python
location_based_events = {}


# near by places within 10 miles radius
radius = 16093
for idx in range(1,len(loc_data.dropna())):
    try:
        tmp_geo = gmaps.geocode(loc_data.loc[idx]['Address'])
    except:
        print("tmp geo code wrong")
    lat = tmp_geo[0]['geometry']['location']['lat']
    lng = tmp_geo[0]['geometry']['location']['lng']
    
    status1, struc1 = biweeklyEvent(lat,lng,radius,'gym',idx)
    status2, struc2 = biweeklyEvent(lat,lng,radius,'art_gallery',idx)
    status3, struc3 = biweeklyEvent(lat,lng,radius,'museum',idx)
    status4, struc4 = biweeklyEvent(lat,lng,radius,'bowling_alley',idx)
    status5, struc5 = biweeklyEvent(lat,lng,radius,'amusement_park',idx)
    if status1 and status2 and status3 and status4 and status5:
        location_based_events.update(struc1)
        location_based_events.update(struc2)
        location_based_events.update(struc3)
        location_based_events.update(struc4)
        location_based_events.update(struc5)
    else:
        print(f"this data has gone wrong: {loc_data.loc[idx]}")
```
